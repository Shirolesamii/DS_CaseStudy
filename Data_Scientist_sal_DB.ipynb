```python
#fetch no.of rows and no.columns,size
print('shape:',db.shape)
print('size:',db.size)
```

```python
#fetch statical summary 
db.describe().T
```

```python
#fetch statical summary of including categorical columns
db.describe(include='object').T
```

```python
#fetch list of unique job titles in Job Title
db['Job Title'].unique()
```

```python
#fetch highes salary record using Avg Salary(K) column
db[db['Avg Salary(K)']==db['Avg Salary(K)'].max()].head(1)
```

```python
#chech how many job availabe for data scientist
db.groupby('Job Title').get_group('Data Scientist').shape[0]
```

```python
#fetch unique company names and  count of unique companys
print(db['Company Name'].unique(),end=" \n\n\n")
print("available company:",db['Company Name'].nunique())
```

```python
# Make a group for salary estimte & show the group wise lower to higher salary estimation
s=db['Salary Estimate'].unique()
s.sort()
print(s)
```

```python
# companies which are having above 4.0 rating
db[db.Rating>=4.0][['Company Name','Rating']]
```

```python
#top 20 locations which are having more no of companies
db['Location'].value_counts().head(20)
```

```python
# fetch only NY loction jobs
db.groupby('Location').get_group('New York, NY')
```

```python
# fetch Data Scientist jobs for NY loction
db.groupby(['Job Title','Location']).get_group(('Data Scientist','New York, NY'))
```

```python
## How many total no of Headquarters are available?
db.Headquarters.nunique()
```

```python
# At which place more no of Headquarters available?
db['Headquarters'].value_counts().head(1)
```

```python
# Fetch top 20 Companies and thier location which have highest no of employees?
db.groupby('Size').get_group('10000+ ')[['Company Name','Location']].head(20)
```

```python
#Newly launched top 20 companies
db.Founded.unique()
db.sort_values('Founded').tail(20)
```

```python
#Which type of ownership have of all companies
db['Type of ownership'].unique()
```

```python
# Fetch data for government companies
db.groupby('Type of ownership').get_group('Government')
```

```python
# How many industry type available
db['Industry'].unique()
```

```python
# I want industrial manufacturing companies
db.groupby('Industry').get_group('Industrial Manufacturing')
```

```python
#Top 10 companies which completed more no of years 
db.sort_values('Age',ascending=False).head(10)
```
